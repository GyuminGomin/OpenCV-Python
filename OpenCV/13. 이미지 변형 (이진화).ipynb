# 13. 이미지 변형 (이진화)

```python
import cv2 # cf. Threshold (= 임계값)

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img, 255//2, 255, cv2.THRESH_BINARY)
# 숫자 2개는 색깔 255를 기준으로 이 공간 사이는 흰색, 밖은 검은색
cv2.imshow('img', img)
cv2.imshow('binary', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Trackbar (값 변화에 따른 변형 확인)

```python
import cv2

def empty(pos) :
    print(pos)
    pass

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow(name)

cv2.createTrackbar('threshold', name, 127, 255, empty)
# bar 이름, 창의 이름, 초기값, 최대값, 이벤트 처리

while True :
    thresh = cv2.getTrackbarPos('threshold', name) # bar 이름, 창의 이름
    ret, binary = cv2.threshold(img, thresh, 255, cv2.THRESH_BINARY)
    
    if not ret :
        break
    
    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q') :
        break
cv2.destroyAllWindows()
```

## Adaptive Threshold
- 이미지를 작은 영역으로 나누어서 임계치 적용

```python
import cv2

def empty(pos) :
    print(pos)
    pass

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow(name)

cv2.createTrackbar('block_size', name, 25, 100, empty) # block사이즈는 홀수만 가느, 1보다는 큰 값
cv2.createTrackbar('c', name, 3, 10, empty) # c는 일반적으로 양수 값을 사용
# bar 이름, 창의 이름, 초기값, 최대값, 이벤트 처리

while True :
    block_size = cv2.getTrackbarPos('block_size', name) # bar 이름, 창의 이름
    c = cv2.getTrackbarPos('c', name)
    
    if block_size <= 1 : # 짝수를 홀수로 바꿔주기 위한 것
        block_size = 3
    if block_size % 2 == 0 :
        block_size += 1
    
    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
    
    cv2.imshow(name, binary)
    if cv2.waitKey(1) == ord('q') :
        break
cv2.destroyAllWindows()
```

## 오츠 알고리즘
Bimodal Image 에 사용하기 적합 (최적의 임계치를 자동으로 발견)
- 모든 이미지에 대해 최적합을 찾지는 못함

```python
import cv2 # cf. Threshold (= 임계값)

img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img, 255//2, 255, cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
# 숫자 2개는 색깔 255를 기준으로 이 공간 사이는 흰색, 밖은 검은색
print('otsu threshold', ret)

cv2.imshow('img', img)
cv2.imshow('binary', binary)
cv2.imshow('otsu', otsu)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
